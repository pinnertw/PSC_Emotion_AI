```python
import matplotlib.pyplot as plt
class_nb = 7
result[result.class_nb == class_nb][['square_mean_valence', 'square_mean_arousal']].plot(kind='box')
plt.xticks([1, 2], ['valence', 'arousal'])
plt.title("Erreur square-mean:")
plt.show()
result[result.class_nb == class_nb][['pearson_val', 'pearson_arousal']].plot(kind='box')
plt.xticks([1, 2], ['valence', 'arousal'])
plt.title("Coefficient de corrÃ©lation Pearson")
plt.show()
for i in result[result.class_nb == class_nb].index:
    plt.scatter(result[result.class_nb == class_nb].iloc[i, -5], result[result.class_nb == class_nb].iloc[i, -3])
plt.xlabel("Valence")
plt.ylabel("Arousal")
plt.title("Erreur square-mean")
plt.show()
for i in result[result.class_nb == class_nb].index:
    plt.scatter(result[result.class_nb == class_nb].iloc[i, -4], result[result.class_nb == class_nb].iloc[i, -2])
plt.xlabel("Valence")
plt.ylabel("Arousal")
plt.title("Coefficient de corrÃ©lation de Pearson")
plt.show()
```
